# Fun with math

Let's start by importing some libraries

```python
# We will use this for plotting
import matplotlib.pyplot as plt
# We will use this for math
import numpy as np
```

```python
x = np.arange(10)
x
```

```python
y = np.arange(10) * 2
y
```

```python
plt.plot(x, y)
```

```python
n = 100
x = np.random.random(n) * 2 - 1
x
```

```python
y = np.random.random(n) * 2 - 1
y
```

```python
plt.scatter(x, y)
```

```python
r = np.sqrt(x**2 + y**2)
r
```

```python
in_circle = r < 1
color = ["red" if hit else "grey" for hit in in_circle]
color
```

```python
plt.scatter(x[in_circle], y[in_circle], c=np.array(color)[in_circle])
```

```python
num_hits = np.sum(in_circle)
num_hits
```

```python
ratio = num_hits / n
ratio
```

```python
ratio * 2 * 2
```
